```python
from collections import defaultdict

def get_stats(text):
    counter = defaultdict(int)

    for pair in zip(text, text[1:]):
        counter[pair] += 1
    
    return counter

stats = get_stats(tokens)
sorted_stats = sorted(((v,k) for k,v in stats.items()) , reverse=True)
```

```python
# 0-255, we will make a new token (256+)

def merge(ids: list[int] , pair: tuple[int,int] , new: int):
    # in the list of ints, replace all pair with new token (new)
    new_ids , i = [] , 0

    while i < len(ids):
        # if we are not at the very last position, AND the pair matches, replace it
        if i < len(ids)-1 and ids[i] == pair[0] and ids[i+1] == pair[1]:
            new_ids.append(new)
            i += 2
        else:
            new_ids.append(ids[i])
            i += 1
    
    return new_ids

print(merge([5,6,6,7,9,1] , (6,7) , 99))  # replace (6,7) with 99
```

```python
def get_stats(ids):
    counter = defaultdict(int)

    for pair in zip(ids, ids[1:]):
        counter[pair] += 1

    return counter

def merge(ids: list[int] , pair: tuple[int,int] , new: int):
    new_ids , i = [] , 0

    while i < len(ids):
        if i < len(ids)-1 and ids[i] == pair[0] and ids[i+1] == pair[1]:
            new_ids.append(new)
            i += 2
        else:
            new_ids.append(ids[i])
            i += 1
    
    return new_ids
```

```python
vocab_size = 272 # desired final vocab size (hyperparameter -> we'll learn the sweet spot later)
num_merges = vocab_size - 256 # number of merges (hyperparameter -> we'll learn the sweet spot later)  # 16 merges

ids = list(tokens) # copying the tokens  # different from ids = [tokens] (which makes a list of lists)

merges = {}  # {int,int} -> int
for i in range(num_merges):
    stats = get_stats(ids)
    pair = max(stats , key=stats.get)
    index = 256 + i

    print(f'merging {pair} into {index}')
    ids = merge(ids , pair , index)

    merges[pair] = index
```

```python
# decoding
# given a sequence of ints in range [0 , vocab_size], what is the text
```

```python
vocab = {index:bytes([index]) for index in range(256)}
# and going up the merge tree
for (p0,p1) , index in merges.items():
    vocab[index] = vocab[p0] + vocab[p1]

def decode(ids: list[int]):
    tokens = b''.join(vocab[index] for index in ids)
    text = tokens.decode('utf-8' , errors='replace')
    return text
```

```python
# encoding

def encode(text: str):
    tokens = list(text.encode('utf-8'))

    while len(tokens) >= 2:
        stats = get_stats(tokens)
        pair = min(stats , key=lambda p: merges.get(p , float('inf')))
        
        if pair not in merges:
            break  # nothing else can be merged

        index = merges[pair]
        tokens = merge(tokens , pair , index)
    
    return tokens


print(encode('Hey There!'))
```
